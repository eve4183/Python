```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Wczytanie danych Titanic
titanic_data = sns.load_dataset('titanic')

# Wyświetlenie kilku pierwszych wierszy danych, aby sprawdzić, czy wczytano poprawnie
print(titanic_data.head())
```

```python
# Wykres słupkowy przedstawiający liczbę pasażerów w poszczególnych klasach podróży w zależności od płci
plt.figure(figsize=(10, 6))

sns.countplot(x='class', hue='sex', data=titanic_data, palette='pastel')

plt.title('Liczba pasażerów w poszczególnych klasach podróży według płci')

plt.xlabel('Klasa podróży')
plt.ylabel('Liczba pasażerów')

plt.show()
```

```python
# Wykres zależności wieku pasażera od ceny biletu z podziałem na płeć
ax = sns.relplot(data=titanic_data, x='age', y='fare', col='sex', alpha=0.7)

# Dodanie tytułu do każdego subplotu
ax.set_titles("Sex: {col_name}")

plt.show()
```

```python
# Wykres zależności wieku pasażera od ceny biletu (fare) z podziałem na płeć  oraz miejsce zaokrętowania (embarked)

ax = sns.relplot(
    data=titanic_data,
    x='age',
    y='fare',
    row='sex',
    col='class',
    hue='embark_town',
    alpha=0.7,
    height=3)

# Dodanie tytułu do każdego subplotu
ax.set_titles("Płeć: {row_name}, Klasa: {col_name}")

plt.show()
```

```python
# Wykres zależności ceny biletu w zależności od klasy i płci pasażera

ax = sns.barplot(
    data=titanic_data,
    x='sex',
    y='fare',
    hue='class',
    ci=None,
    palette='hot')

plt.title('Cena biletu w zależności od płci i klasy')

plt.show()
```

```python
# Wykres cena biletu w zależności od kategorii pasażera i klasy


#  Własna paleta kolorów
moja_paleta = {'First': '#FF7F50', 'Second': '#20B2AA', 'Third': '#90EE90'}

ax = sns.barplot(
    data=titanic_data,
    x='who',
    y='fare',
    hue='class',
    palette=moja_paleta,
    ci=None,
    edgecolor='mediumblue')

plt.title('Cena biletu w zależności od kategorii pasażera i klasy')

plt.xlabel('Kategoria pasażera')
plt.ylabel('Cena biletu')

plt.show()
```

```python
# Wykres liniowy - Cena biletu w zależności od wieku pasażera i klasy

sns.lineplot(data=titanic_data, x='age', y='fare', hue='class')

plt.title('Cena biletu w zależności od wieku pasażera i klasy')

plt.xlabel('Wiek pasażera')
plt.ylabel('Cena biletu')

plt.show()
```

```python
# Wykres boxplot - Cena biletu w zależności od klasy podróży

sns.boxplot(x='class', y='fare', data=titanic_data, palette='Set3')

plt.title('Cena biletu w zależności od klasy podróży')

plt.xlabel('Klasa podróży')
plt.ylabel('Cena biletu')

plt.show()
```

```python
# Wykres słupkowy przedstawiający przeżywalność w zależności od płci i miejsca, z którego wsiadali

sns.barplot(x='sex', y='survived', hue='embark_town', data=titanic_data, palette='pastel')

plt.title('Przeżywalność pasażerów w zależności od płci i miejsca, z którego wsiadali')

plt.xlabel('Płeć')
plt.ylabel('Przeżywalność')

plt.legend(title='Miejsce embarked', loc='upper right')

plt.show()
```

```python
# Wykresy przeżywalności

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Przeżywalność na Titanicu według różnych kryteriów')

# Pierwszy wykres - przeżywalność według płci
sns.barplot(ax=axes[0], x='sex', y='survived', data=titanic_data, palette='Set1')
axes[0].set_title("Płeć")

# Drugi wykres - przeżywalność według klasy podróży
sns.barplot(ax=axes[1], x='class', y='survived', data=titanic_data, palette='Set1')
axes[1].set_title("Klasa podróży")

# Trzeci wykres - przeżywalność według miejsca wsiadania
sns.barplot(ax=axes[2], x='embark_town', y='survived', data=titanic_data, palette='Set1')
axes[2].set_title("Miejsce wsiadania")

plt.show()
```

```python
# Rozkład wieku pasażerów w zależności od klasy podróżysns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.histplot(data=titanic_data, x='age', hue='pclass', kde=True, multiple='stack', palette='husl')

plt.title('Rozkład wieku pasażerów w zależności od klasy podróży')
plt.xlabel('Wiek')
plt.ylabel('Liczba pasażerów')

plt.show()
```
