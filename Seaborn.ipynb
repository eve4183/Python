# **Wykres liniowy**

```python
y_1 = [random.randint(-100,100) for x in range(50)]
y_2 = [random.randint(-100,100) for x in range(50)]
y_3 = [random.randint(-100,100) for x in range(50)]

x = [x for x in range(50,100)]

y_1 = np.cumsum(y_1)
y_2 = np.cumsum(y_2)
y_3 = np.cumsum(y_3)
```

```python
sns.relplot(x=x, y=y_1)
```

```python
import numpy as np
import pandas as pd

# Wygeneruj dane losowe dla akcji
akcje = np.random.randint(-100, 100, size=(20, 3))

# Wykonaj kumulacyjną sumę dla każdej kolumny
akcje = np.cumsum(akcje, axis=0)

# Utwórz DataFrame
df = pd.DataFrame(akcje, columns=['Akcja1', 'Akcja2', 'Akcja3'])

# Dodaj kolumnę 'x' z indeksami
df['x'] = df.index

df.head
```

```python
sns.relplot(data=df, x=df['x'], y='Akcja1', kind = 'line')
sns.relplot(data=df, x=df['x'], y='Akcja2', kind = 'line')
```
