```python
from faker import Faker
import random
fake = Faker()
#wprowadzanie 10 wygenerowanych sztucznie wierszy
for i in range (1,101):
  rowkey = "e"+str(i)
  row = table2.direct_row(rowkey)

  fake = Faker()

# tytuł filmu, rok produkcji
  row.set_cell("film_info","tytul",fake.sentence())
  row.set_cell("film_info","year",fake.year())

# klasyfikacja wiekowa
  age_values = [0, 7, 16, 18]
  age_value = str(random.choice(age_values))
  row.set_cell("film_info", "age", age_value)

# ocena IMDb
  imdb_value = str(fake.random_int(min=1, max=10))
  row.set_cell("film_info", "imdb", imdb_value)

# ocena Rotten_Tomatoes
  rt_value = str(fake.random_int(min=1, max=10))
  row.set_cell("film_info", "rotten_tomatoes", rt_value)

# dostępność filmu na danej platformie streamingowej
  row.set_cell("platforma","disney", "yes" if random.choice([0, 1]) == 1 else "no")
  row.set_cell("platforma","netflix", "yes" if random.choice([0, 1]) == 1 else "no")
  row.set_cell("platforma","hulu", "yes" if random.choice([0, 1]) == 1 else "no")

  row.commit()
```

```python
# Usuń kolumnę 'hulu'

column_family = b"platforma"
column_qualifier = b"hulu"

# Pobierz wszystkie wiersze z tabeli
rows = table2.read_rows()
for row in rows:
    print(f"Row key: {row.row_key.decode('utf-8')}")
    for family, columns in row.cells.items():
      for column, cells in columns.items():
        for cell in cells:
          print(family,":",column.decode('utf-8'),"=>",cell.value.decode('utf-8'),cell.timestamp)

# Iteruj przez wiersze
for row_key, row in rows:
    # Sprawdź, czy kolumna "hulu" istnieje w rodzinie kolumn "platforma"
    if column_family in row.cells and column_qualifier in row.cells[column_family]:
        # Usuń kolumnę "hulu"
        del row.cells[column_family][column_qualifier]

        row.commit()

print("Usunięto kolumnę 'hulu' dla wszystkich wierszy.")
```

```python
# dodanie grupy kolumn 'twórcy' i kolumny 'reżyser'

# Pobierz wszystkie wiersze z tabeli
rows = table2.read_rows()
for row in rows:
    print(f"Row key: {row.row_key.decode('utf-8')}")
    for family, columns in row.cells.items():
      for column, cells in columns.items():
        for cell in cells:
          print(family,":",column.decode('utf-8'),"=>",cell.value.decode('utf-8'),cell.timestamp)

# Iteruj przez wiersze
for row_key, row in rows:
    # Ustaw kolumnę "rezyser" w grupie kolumn "tworcy" dla każdego wiersza
    row.set_cell("tworcy", "rezyser")

    # Zapisz zmiany w wierszu
    row.commit()

print("Dodano kolumnę 'rezyser' do grupy kolumn 'tworcy' dla wszystkich wierszy.")
```
