```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
y_1 = [random.randint(-100,100) for x in range(50)]
y_2 = [random.randint(-100,100) for x in range(50)]
y_3 = [random.randint(-100,100) for x in range(50)]

x = [x for x in range(50,100)]

y_1 = np.cumsum(y_1)
y_2 = np.cumsum(y_2)
y_3 = np.cumsum(y_3)

plt.plot(y_1)
plt.plot(y_2)
plt.plot(y_3)

plt.show()
```

```python
# kolor

plt.figure(figsize = (20,7))
plt.plot(x, y_1, color = 'red')
```

```python
# marker

plt.figure(figsize = (20,7))
plt.plot(x, y_1, marker = 'D')
```

```python
# styl linii

plt.figure(figsize = (20,7))
plt.plot(x, y_1, linestyle = 'dashed', color = 'red', marker = 'D', linewidth = 2)
```

```python
# fmt (linia, marker, kolor)
# - linia, X marker, g green
plt.figure(figsize = (20,7))
plt.plot(x, y_1,'-Xg')
```

```python
# Zadanie 1. Narysowac 3 wykresy

plt.figure(figsize = (20,7))
plt.plot(x, y_1, '-Pg', linewidth = 1)
plt.plot(x, y_2, linestyle = 'dotted', color = 'magenta', marker = '*', linewidth = 1)
plt.plot(x, y_3, linestyle = 'dashed', color = 'blue', marker = 'D', linewidth = 1)

plt.show()
```

```python
# Subplot - porównanie wykresów obok siebie. (1 - liczba wykresow w osi x, 3 - w osi y, 1 - numer wykresu)

plt.subplot(1, 3, 1)
plt.plot(x, y_1, '-Pg')

plt.subplot(1, 3, 2)
plt.plot(x, y_2, '-Xr')

plt.subplot(1, 3, 3)
plt.plot(x, y_3, '-Pb')
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, y_1, '-Pg')

plt.subplot(3, 1, 2)
plt.plot(x, y_2, '-Xr')

plt.subplot(3, 1, 3)
plt.plot(x, y_3, '-Pb')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y_1, '-Pg')

plt.subplot(2, 2, 2)
plt.plot(x, y_2, '-Xr')

plt.subplot(2, 2, 3)
plt.plot(x, y_3, '-Pb')
```

```python
# tworzenie siatki. axs wskazuje pozycje, w ktorej chcemy dany wykres umiescic

fig, axs = plt.subplots(1,3)

axs[0].plot(x, y_1, '-Pg')
axs[1].plot(x, y_2, '-Xr')
axs[2].plot(x, y_3, '-Pb')
```

```python
fig, axs = plt.subplots(3,1)

axs[0].plot(x, y_1, '-Pg')
axs[1].plot(x, y_2, '-Xr')
axs[2].plot(x, y_3, '-Pb')
```

```python
fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, y_1, '-Pg')
axs[0,1].plot(x, y_2, '-Xr')
axs[1,1].plot(x, y_3, '-Pb')
```

```python
import numpy as np
import pandas as pd

# Wygeneruj dane losowe dla akcji
akcje = np.random.randint(-100, 100, size=(20, 3))

# Wykonaj kumulacyjną sumę dla każdej kolumny
akcje = np.cumsum(akcje, axis=0)

# Utwórz DataFrame
df = pd.DataFrame(akcje, columns=['Akcja1', 'Akcja2', 'Akcja3'])

# Dodaj kolumnę 'x' z indeksami
df['x'] = df.index

df.head
```

```python
plt.plot('Akcja1', data= df)

plt.xlabel(xlabel = 'nr dnia')
plt.ylabel(ylabel = 'wskaźnik ceny')
plt.xlim(-5, 20)
plt.ylim(0, 400)
```

```python
plt.xticks([2,4,6,8], labels = ['dwa', 'cztery', 'szesc', 'osiem'])
plt.plot('Akcja1', data= df)
plt.plot('Akcja2', data= df)
plt.xticks(rotation = 45)
```
